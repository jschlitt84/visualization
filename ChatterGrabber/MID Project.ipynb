### Hourly histogram function

```python
def chart(occurance_list,title):
    hour_list = [t.hour for t in occurance_list]
    numbers=[x for x in xrange(0,23)]
    labels=map(lambda x: str(x), numbers)
    plt.xticks(numbers, labels)
    plt.xlim(0,23)
    plt.title(title,size = 14)
    plt.xlabel("Hour")
    plt.ylabel("Tweets")
    plt.hist(hour_list,bins=23)
    plt.show()
```

### Plot distribution by day and time

```python
def dailyDistributionPlot(collected,titles,bigTitle):
    """ draw the histogram of the daily distribution """
    ax = plt.figure(figsize=(18,5)).gca()
    merged = []
    prepped = []
    for ytime in collected:
        numtime = [date2num(t) for t in ytime] 
        merged += numtime
        prepped.append(numtime)
        
    first = num2date(min(merged))
    last = num2date(max(merged))
    
    hour = timedelta(hours=1) 
    minute = timedelta(minutes=1)
    
    firstRounded = first.replace(second=0,minute=0)
    lastRounded = last.replace(second=0,minute=0) + hour
    
    plotMin = date2num(firstRounded)
    plotMax = date2num(lastRounded)
    
    difference = (lastRounded-firstRounded)
    hourBlocks = difference.days*24 + difference.seconds//3600
    if hourBlocks%2 != 0:
        plotMax = date2num(lastRounded + hour)
        hourBlocks += 1
    
    for i in range(len(prepped)):
        numtime = prepped[i]
        plotTitle = titles[i]
        _, _, patches = plt.hist(numtime, bins=(hourBlocks)/2, alpha=0.5, 
                                    stacked=False, range=(plotMin,plotMax))
        
    tks = [num2date(p.get_x())+minute for p in patches] 
    plt.xticks(tks,rotation=90,size = 8)
    plt.xlabel("Date/time")
    plt.gca().set_xlim([plotMin,plotMax])
    plt.ylabel("Number of tweets")
    plt.legend(titles,"upper left")
    plt.title(bigTitle,size=16)
    ax.xaxis.set_major_formatter(DateFormatter('%m/%d %H'))
```

### Chicago and national distribution by date * time

```python
dailyDistributionPlot([firstHandDates,mediaDiscussionDates],['First hand reports','Media discussion'],
                      "National First Hand and Media Related Gun Violence Tweets from 4/18/2014-4/29/2014")
dailyDistributionPlot([chiFirstHandDates,chiMediaDiscussionDates],['First hand reports','Media discussion'],
                      "Chicago First Hand and Media Related Gun Violence Tweets from 4/18/2014-4/29/2014")
```
