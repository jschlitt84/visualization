### Inline video animation

```python
#Animating via tutorial at:
#http://jakevdp.github.io/blog/2013/05/12/embedding-matplotlib-animations/

VIDEO_TAG = """<video width="800" height="600" controls>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Your browser does not support the video tag.
</video>"""

def display_animation(anim,fileName):
    print "Preparing to render video in notebook..."
    plt.close(anim._fig)
    return HTML(anim_to_html(anim,fileName))

def anim_to_html(anim,fileName):
    fileName = fileName+'.mp4'
    video = open(fileName, "rb").read()
    anim._encoded_video = video.encode("base64")
    
    return VIDEO_TAG.format(anim._encoded_video)
```

### Generating geographic data subsets

```python
#Change names here!
categories = {'0':'Irrelevant',
              '1':'Cat1',
              '2':'Cat2',
              '3':'Cat3',
              '4':'Cat4',
              '5':'Cat5'}

cat1Vacc = getCatSub(vaccData,'1','null')
cat2Vacc = getCatSub(vaccData,'2',categories)
cat3Vacc = getCatSub(vaccData,'3',categories)
cat4Vacc = getCatSub(vaccData,'4',categories)
cat5Vacc = getCatSub(vaccData,'5',categories)
listed = [cat1Vacc,cat2Vacc,cat3Vacc,cat4Vacc,cat5Vacc]
for entry in listed:
    print "Length: %s\t Category: %s" % (len(entry['data']),entry['name'])
```

### Return list of subsets

```python
def getListSub(listed,sought,exclusions,prefix,field):
    newList = []
    for entry in listed:
        newData = getFieldSub(entry,sought,exclusions,prefix,field)
        if len(newData['data']) > 1:
            newList.append(newData)
    return newList
```

### Make Pie Chart

```python
def tweetyPie(listed,title,stripText = 'Ebola Tracker',startangle=180,explode = .1):
    fig = plt.figure(1, figsize=(6,6))
    sizes = []; names = []
    for entry in listed:
        sizes.append(len(entry['data']))
        names.append(entry['name'].replace(stripText,''))
                     
    plt.pie(sizes, labels=names, autopct='%1.1f%%', 
            shadow=True, startangle=startangle, explode = [explode]*len(listed))

    plt.title(title)
    return plt
    plt.show()
```

### Analysis per listed subsets

```python
def runAll(listed,
           title,
           startangle=180,
           timeShift=0,stacked=True,
           show=True,
           stripText = ' Ebola Tracker',
           explode = 0.1):
    listed = [entry for entry in listed if len(entry['data']) != 0]
    names = [entry['name'].replace(stripText,'') for entry in listed]
    
    tweetyPie(listed,title,startangle = 180, stripText = stripText, explode = explode)
    fig = dailyDistributionPlot(listed,names,title,timeShift, show = False)
    fig.show()
    
    groupHourly(listed, names, title + ' by hour (utc)', timeShift, stacked=stacked,show=True)
    groupDaily(listed, names, title + ' by day (utc)', timeShift, stacked=stacked,show=True)
   
    print; print
    for entry in listed:
        shortName = entry['name'].replace(stripText,'')
        
        print "Data and analysis for", shortName
        print "Random Tweet Sample:"
        printSample(entry['data'],shortName,5)
        print "Top 5 Hashtags:"
        countHashTags(entry['data'],number = 5)
        
        subsetCats = sorted(list(set(entry['data']['NLPCat'])))
        catSubs = []
        for cat in subsetCats:
            catSubs.append(getFieldSub(entry,[cat],[],'','NLPCat'))
        fig = dailyDistributionPlot(catSubs,subsetCats,shortName+'NLP Tweets',timeShift,show = False)
        fig.show()                   
        
        try:
            clouds = getWordWeights(entry,999,'','')
            showWordCloud(clouds['all'],title="Wordcloud for all data for %s" % shortName)
        except:
            print "Could not generate WordCloud, sample too small"
        mapSubject(entry,shortName, background = 'blue marble')
```

### Overall tweet volume from listed countries

```python
runAll(listed,'Tweets By Category', stacked = True)
```
